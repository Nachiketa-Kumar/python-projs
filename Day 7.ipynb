## Project 15: Hangman Project 
#### If the word maches the needed letter in the word then length will not increase and if it length becomes max then you loose.

```python
import random
word_list=["ardvark", "baboon", "camel", "zoey"]
print(word_list)
s=input("Enter any word of length 7 you want to add in the list above ----> ")
word_list.append(s)
print("new list is ",word_list)
chosen_word=random.choice(word_list)
l#letter=input("guess a letter: ")

Hang=[''' _______
         |/      |
         |      
         |      
         |       
         |       
         |
      ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |      
     |       
     |       
     |
  ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |       |
     |       
     |       
     |
  ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |      \|/
     |       
     |       
     |
  ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |       
     |
  ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |      / 
     |
  ___|___''',
      '''
       _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |      / \
     |
  ___|___''']
end_of_game=False
d=[]
lives=len(chosen_word)
for i in range(len(chosen_word)):
    d+="_"
print(d)
lives=len(Hang)
print(f"lifes left {lives}")
while(not end_of_game):   
    letter=input("guess a letter: ")
    for j in range(len(chosen_word)):
        l=chosen_word[j]
        if(l==letter):
            d[j]=l
        """else:
            if(lives==0):
                print("Game Over")
                break
            lives=lives-1"""
    if(letter not in chosen_word):
        lives-=1
        if(lives==0):
            end_of_game=True
            print("Game Over, You Loose!")
            
        
    print(*list(d))
    print(letter+" is the guessd letter.")
    if("_" not in d):
        end_of_game=True
        print("Congrats, You Win!")
    print(Hang[lives])
print(chosen_word+" was the hangman's word.")
```
